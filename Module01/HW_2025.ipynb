```python
# Q6. Median value of horsepower
# Find the median value of the horsepower column in the dataset.
# Next, calculate the most frequent value of the same horsepower column.
# Use the fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step.
# Now, calculate the median value of horsepower once again.
cars.horsepower
```

```python
median_before = cars.horsepower.median()
print(median_before)
```

```python
most_freq = cars.horsepower.mode()[0]
print(most_freq)
```

```python
cars.horsepower = cars.horsepower.fillna(most_freq)
print(cars.horsepower)
```

```python
median_after = cars.horsepower.median()
print(median_after)
```

```python
# Q7. Sum of weights
asian_car = cars[cars['origin'] == 'Asia']
print(asian_car)
```

```python
sel_asian_car = asian_car[['vehicle_weight', 'model_year']].head(7)
sel_asian_car 
```

```python
X = sel_asian_car.values
X
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = XTX_inv.dot(X.T).dot(y)
w
```

```python
w.sum()
```
